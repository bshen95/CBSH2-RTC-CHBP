# Figure2 Cactus plots (Runtime)

```python
graph_name = "random-32-32-20"
time_out = 60
scenario ="merge"
data_length=50
N =  pd.DataFrame()
BP =  pd.DataFrame()
CH =  pd.DataFrame()
CHBP =  pd.DataFrame()
column_name = "runtime"
for agent_size in range(20, 80, 10):
    df1 = load_data("../results/"+graph_name+"/"+scenario+"/agent-"+str(agent_size)+"-"+str(time_out)+"-N.csv")
    N = N.append(df1,ignore_index=True)
    
    df2 = load_data("../results/"+graph_name+"/"+scenario+"/agent-"+str(agent_size)+"-"+str(time_out)+"-BP.csv")
    BP = BP.append(df2,ignore_index=True)  
    
    df3 = load_data("../results/"+graph_name+"/"+scenario+"/agent-"+str(agent_size)+"-"+str(time_out)+"-CH.csv")
    CH = CH.append(df3,ignore_index=True)
    
    df4 = load_data("../results/"+graph_name+"/"+scenario+"/agent-"+str(agent_size)+"-"+str(time_out)+"-CHBP.csv")
    CHBP = CHBP.append(df4,ignore_index=True)

# set all unsolvable instances to 60 sec 
x = list(range(0, 300))
N = N[N['solution cost'] > 0 ]
x_n = list(range(0, len(N)))
BP = BP[BP['solution cost'] > 0 ]
x_bp = list(range(0, len(BP)))
CH = CH[CH['solution cost'] > 0 ]
x_ch = list(range(0, len(CH)))
CHBP = CHBP[CHBP['solution cost'] > 0 ]
x_chbp = list(range(0, len(CHBP)))

# sort instances by runtime 
N = N.sort_values(by=[column_name],ignore_index=True)
BP = BP.sort_values(by=[column_name],ignore_index=True)
CH = CH.sort_values(by=[column_name],ignore_index=True)
CHBP = CHBP.sort_values(by=[column_name],ignore_index=True)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x_n,  list(N[column_name]), linestyle ='-.', color='C0', label='WDG',linewidth=2,zorder=2)
ax.plot(x_bp, list(BP[column_name]), linestyle =(0, (1, 1)), color='C1',label='WDG+BP', linewidth=2,zorder=0)
ax.plot(x_ch, list(CH[column_name]), linestyle =(0, (5, 1)), color='C2',label='WDG+CH', linewidth=2,zorder=1)
ax.plot(x_chbp, list(CHBP[column_name]), linestyle ='-', color='C3',label='WDG+CHBP', linewidth=2,zorder=3)
ax.set_ylabel('Runtime (sec)',weight ='bold',size ='14',labelpad = 0)
ax.set_xlabel('#Solved Instances',weight ='bold',size ='14',labelpad = 2)
plt.title('Random Map',fontweight="bold",size ='16')
plt.xticks(fontsize= 14)
plt.yticks(fontsize= 14)
ax.legend(loc='lower right',fancybox=True, framealpha=0,ncol=2, prop={'size': 12})
ax.set_yscale('log')
ax.set_ylim([0.001,100])


# add background image 
im = mpimg.imread('random-32-32-20.png')
ax2 = fig.add_axes([0.15, 0.56, 0.3, 0.3], anchor='SE', zorder=3)
ax2.imshow(im)
ax2.axis('off')
ax.grid(True,which='both',linestyle='-', linewidth=0.01,color ='gainsboro',zorder=-1)
plt.show()
fig.savefig("fig/random_"+column_name+"_distribution.pdf", bbox_inches='tight')
```
